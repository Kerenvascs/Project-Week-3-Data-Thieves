```python
import pandas as pd
import numpy as np
import requests
from pandas.io.json import json_normalize
from flatten_json import flatten
from datetime import date
```

# OpenAQ API Conection

## List Countries

```python
url = "https://api.openaq.org/v1/countries"

response = requests.get(url)
response
```

```python
response.json()
# response.text
response.json()['results']
```

```python
# Convert to DF
data = pd.DataFrame(response.json()['results'])
data['name'].unique()
data.head()
# data.columns
```

```python
# Countries with the highest number of observations
data = pd.DataFrame(json_normalize(response.json()['results']))
data.sort_values(by='count', ascending=False).head(20)
```

## Listing cities

```python
# Cities
url = "https://api.openaq.org/v1/cities"

response = requests.get(url)
response
```

```python
response.json()
```

```python
cities = pd.DataFrame(json_normalize(response.json()['results']))
cities.sort_values(by='count', ascending=False)
```
