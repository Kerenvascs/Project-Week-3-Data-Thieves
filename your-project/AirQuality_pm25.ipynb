# Requesting data

## Bogota - Colombia

```python
# Colombia
begin = date(2020,3,1)
end = date(2020,3,31)

api_string = "https://api.openaq.org/v1/measurements?location={location}&parameter={measure}&date_from={fromdate}&date_to={todate}&limit=5000"
location = "US Diplomatic Post: Bogota"
measure = "pm25"

response = requests.get(api_string.format(location=location,measure=measure,fromdate=begin,todate=end)).json()
response

dic_flatten = [flatten(d) for d in response['results']]
dic_flatten
bogota_2003 = pd.DataFrame(dic_flatten).reset_index()
bogota_2003.head()
```

```python
# criar coluna março e fevereiro encoded.
data['date'] = pd.to_datetime(data['date_local']) 
data['date_1'] = data['date'].values.astype('datetime64[M]')
data['month'] = data['date_1'].dt.month 
data.head()

# Transformar em str e pegar os dois ultimos
```

```python
# Removing columns
data.columns
data.drop(columns=['Unnamed: 0'], inplace =True)
data.columns
```

```python
# Cleaning names
data['city'].replace(['New York-Northern New Jersey-Long Island'], 'New York', inplace = True)
data.city.unique()
```

```python
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns
msno.matrix(data)
```

```python
data.isnull().sum()
```

```python
# remove rows <400
data = data.where(data['value'] < 400)
data.sort_values(by="value").head()
data = data.dropna()
# data.shape (10395)

# value  > -400
data = data.where(data['value'] > -100)
data.sort_values(by="value").head()
data = data.dropna()
data.shape
```

```python
pm25 = data.groupby(['city', 'month'])['value'].agg(['mean', 'min', 'max','median']).reset_index()
pm25.head(8)
# data.groupby(['city', 'month'])['value'].describe()
```
